```python
import os
import shutil
import pandas as pd

# Load the DataFrame
bg_dats = pd.read_csv('/Users/ashleywalton/Dropbox/1_BrainModulationLab/0_RNS_Processing_and_Analysis_2023/5a_Analysis_of_BG_recordings/1_BG_Dats_Ft_Summary_connect_to_atlas_label/June_2024/bg_dats_per_chan_center_labeled.csv')

def copy_dat_files(rns_ids, base_path, out_path, bg_dats):
    for rns_id in rns_ids:
        # Create the output directory for this rns_id if it doesn't exist
        out_dir = os.path.join(out_path, rns_id)
        os.makedirs(out_dir, exist_ok=True)

        # Filter the dataframe for the current rns_id and convert filenames to strings
        rns_files = bg_dats[bg_dats['rns_id'] == rns_id]['dat_file_nm'].astype(str).tolist()

        # Define the path to the directory containing the dat files for this rns_id
        rns_dir = os.path.join(base_path, rns_id)

        if not os.path.exists(rns_dir):
            print(f"Directory not found: {rns_dir}")
            continue

        # Iterate through the list of dat files and copy them to the output directory
        for file_name in rns_files:
            # Add the ".dat" extension if it's not already present
            if not file_name.endswith(".dat"):
                file_name += ".dat"

            source_file = os.path.join(rns_dir, file_name)
            if os.path.exists(source_file):
                shutil.copy(source_file, out_dir)
                print(f"Copied {file_name} to {out_dir}")
            else:
                print(f"File not found: {source_file}")

# Example usage:
# Define your list of rns_ids
rns_ids = ["RNS1384","RNS3455","RNS6786","RNS1457","RNS8966","RNS9218","RNS9943"]

# Define the base path where the folders are located
base_path = "/Volumes/Nexus/RNS_data/ecog_reconstructed/MGH/"

# Define the output path where the files will be copied to
out_path = "/Users/ashleywalton/Dropbox/1_BrainModulationLab/0_RNS_Processing_and_Analysis_2023/5a_Analysis_of_BG_recordings/3_BG_Dats_Waveform_Analysis/selected_dats"

# Call the function
copy_dat_files(rns_ids, base_path, out_path, bg_dats)
```
